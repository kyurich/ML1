```python
#удаляем столбцы, где отсутствуют все значения
df = dropNA(df, axis = 'columns', how = 'all')
df.head()
```

```python
df = replace(df, ['MinTemp', 'MaxTemp', 'Rainfall', 'WindGustSpeed', 'WindSpeed9am', 'Humidity9am', 'Pressure9am', 'Pressure3pm'],  value = 'mean')
```

```python
df.head()
```

```python
df = df.drop(columns= ['WindGustDir', 'WindDir9am'])
```

```python
array = df.values
X = array[:, 3:9]
Y = array[:, 2]
validation_size = 0.2
seed = 7
X_train, X_validation, Y_train, Y_validation = model_selection.train_test_split(X, Y, test_size=validation_size, random_state=seed)
```

```python
# Make predictions 
LR = LinearRegression()
LR.fit(X_train, Y_train)
predictions = LR.predict(X_validation)
print(LR.score(X_train, Y_train))
print(LR.coef_)
print(LR.intercept_ )
```
