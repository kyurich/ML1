```python
from sklearn.model_selection import train_test_split
seed = 7
X_train, X_test, y_train, y_test = train_test_split(np.array(df.Rainfall).reshape(-1,1), np.array(df.MaxTemp).reshape(-1,1), test_size=0.2, random_state=7)
```

```python
models = []

#models.append(('LDA', LinearRegression()))

models.append(('CART', DecisionTreeClassifier()))

models.append(('SVM', SVC()))
# evaluate each model in turn
results = []
names = []
for name, model in models:
    kfold = model_selection.KFold(n_splits=10, random_state=seed)
    cv_results = model_selection.cross_val_score(model, X_train.astype(int), y_train.astype(int), cv=kfold, scoring='accuracy')
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
```
